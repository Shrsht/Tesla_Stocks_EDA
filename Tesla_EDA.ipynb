```python
import pandas as pd
import yfinance as yh
import datetime as dt
import sklearn 
#from pandas_datareader import data as pdr


import plotly.offline as pt
pt.init_notebook_mode(connected = True)

import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio

import mplfinance as mpf
```

```python
end = dt.datetime.now()
start = end - dt.timedelta(days = 365) #1 year ago
```

```python
Tesla_df = yh.Ticker("TSLA").history(start = start,end = end)

# Tesla_df['Date'] = Tesla_df['Date'].dt.date
# Tesla_df = Tesla_df.set_index('Date')

#Tesla_df = Tesla_df.set_index(Tesla_df.index.date)              #,'Stock Splits'])
# Tesla_df = Tesla_df.set_index('Date')
Tesla_df['Tomorrow'] = Tesla_df['Close'].shift(-1)
Tesla_df['Direction'] = pd.Series(Tesla_df['Close'] > Tesla_df['Tomorrow'])
Tesla_df['Direction'] = Tesla_df['Direction'].astype(int)
Tesla_df = Tesla_df.drop(columns = ['Dividends','Stock Splits'])
Tesla_df
```

```python
pd.options.plotting.backend = 'plotly'
Tesla_df.plot()
```

# CANDLESTICK GRAPHS:


```python
Tesla_df['10day_avg'] = Tesla_df['Close'].rolling(window = 10, min_periods = 0).mean()
Tesla_df['30day_avg'] = Tesla_df['Close'].rolling(window = 30, min_periods = 0).mean()
Tesla_df['50day_avg'] = Tesla_df['Close'].rolling(window = 50, min_periods = 0).mean()
```

```python
#go.Candlestick(x = Tesla_df.index, open = Tesla_df['Open'], high = Tesla_df['High'],low = Tesla_df['Low'], name = "OHLC")
```

```python
fig = make_subplots(rows = 2, cols = 1 , shared_xaxes =True, vertical_spacing = 0.1, 
                    subplot_titles = ('TSLA',"Volume"), row_width = [0.2,0.7])

fig.add_trace(go.Candlestick(x = Tesla_df.index, open = Tesla_df['Open'], high = Tesla_df['High'],
                             low = Tesla_df['Low'], close = Tesla_df['Close'], name = "OHLC"),row=1,col=1)
```

```python
fig.add_trace(go.Scatter(x = Tesla_df.index, y = Tesla_df['50day_avg'],marker_color = 'grey', name = "MA50" ))
fig.add_trace(go.Scatter(x = Tesla_df.index, y = Tesla_df['10day_avg'],marker_color = 'blue', name = "MA10" ))
```

```python
fig.add_trace(go.Bar(x = Tesla_df.index , y = Tesla_df['Volume'], marker_color = 'blue',showlegend = False),row = 2,
              col = 1)
```
